```python
import numpy as np
import pandas as pd 
```

```python
def f_linear(x, surface_val, coef):

    return coef[0]*x*surface_val + surface_val*coef[1]


def f_polynom2(x, surface_val, coef):

    return coef[0]*surface_val*x**2 + coef[1]*surface_val*x + surface_val*coef[2]


def f_polynom3(x, surface_val, coef):

    return coef[0]*surface_val*x**3 + coef[1]*surface_val*x**2 + coef[2]*surface_val*x + surface_val*coef[3]


function_dict = {'linear': f_linear,
                 'polynom2': f_polynom2, 'polynom3': f_polynom3}
```

```python
trend = ['none', 'linear', 'nonlinear', 'random_con', 'random_dis']
relationship = ['linear', 'polynom2', 'polynom3']
coeficients = [[[1.5, 0.25], [1.5, -0.25], [-1.5, 0.25], [-1.5, -0.25]],
               [[2, -1, 0.25], [-2, -1, -0.25], [-2, 1, -0.25], [-2, 1, 0.25]],
               [[-3, -2, 2, 0.25], [3, -2, 2, -0.25], [-3, 2, -2, 0.25], [-3, -2, 2, -0.25]]]
surface_level = ['small', 'medium', 'large']
spatial_autocorr = ['high', 'low', 'none']

trend_list = []
relationship_list = []
dataset_path_list = []
target_path_list = []
trend_path_list = []
coeficients_list = []
spatial_autocorr_list = []
surface_level_list = []
id_list = []
id  = 0

for x_n, x in enumerate(trend):
    for w_n, w in enumerate(surface_level):
        for y_n, y in enumerate(relationship):
            for z_n, z in enumerate(spatial_autocorr):

                trend_list.append(x)
                relationship_list.append(y)
                spatial_autocorr_list.append(z)
                surface_level_list.append(w)
                dataset_path_list.append(f'dataset-TR_{x}_rel-{y}-lvl_{w}-sac-{z}.csv')
                target_path_list.append(f'target-TR_{x}_rel-{y}-lvl_{w}-sac-{z}.csv')
                trend_path_list.append(f'trend_{x}.csv')
                coeficients_list.append(coeficients[y_n])
                id_list.append(id)
                id += 1

dataset = pd.DataFrame(data={'id':id_list,
                            'relationship': relationship_list,
                            'coeficients': coeficients_list,
                            'trend': trend_list,
                            'surface_level':surface_level_list,
                            'spatial_autocorrelation':spatial_autocorr_list,
                            'dataset_path': dataset_path_list,
                            'target_path': target_path_list,
                            'trend_path': trend_path_list})
dataset
```

```python
dataset.to_csv('dataset_info.csv', index=False)
```
